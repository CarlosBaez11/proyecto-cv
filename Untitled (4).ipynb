```python
datamodule = UCSDped(
    root="./datasets/ucsdped",
    augmentations=augmentations,
    # clip_length_in_frames=CLIP_LEN,
)
```

```python
import torch
import matplotlib.pyplot as plt
import numpy as np

datamodule.setup(stage="test")
model = engine.model
model.eval()

# Obtener un clip del test set (T, C, H, W)
test_loader = datamodule.test_dataloader()
sample = next(iter(test_loader))
clip = sample["image"][0].to(model.device)  # (T, C, H, W)

# Tomar el frame central del clip
frame = clip[clip.shape[0] // 2]

# Inferencia
with torch.no_grad():
    outputs = model(clip.unsqueeze(0))
```

```python

# Extraer mapas
anomaly_map = outputs.anomaly_map.squeeze().cpu().numpy()
pred_mask = outputs.pred_mask.squeeze().cpu().numpy()
pred_score = outputs.pred_score.item()
pred_label = outputs.pred_label.item()

# Normalizar mapa
anomaly_map = (anomaly_map - anomaly_map.min()) / (anomaly_map.max() - anomaly_map.min() + 1e-8)

# Convertir el frame a numpy (H, W, C)
frame_np = frame.cpu().permute(1, 2, 0).numpy()
```

```python


# Visualizar
plt.figure(figsize=(14, 4))

plt.subplot(1, 3, 1)
plt.imshow(frame_np)
plt.title("Frame central del clip")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(frame_np)
plt.imshow(anomaly_map, cmap="jet", alpha=0.5)
plt.title(f"Mapa de anomalía\nscore={pred_score:.3f}, label={pred_label}")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(pred_mask, cmap="gray")
plt.title("Máscara umbralizada")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# Clip con anomalia
clip = None
outputs = None

for batch in datamodule.test_dataloader():
    with torch.no_grad():
        preds = model(batch["image"].to(model.device))
    
    # preds.pred_label tiene shape (batch_size,)
    for i in range(len(preds.pred_label)):
        if preds.pred_label[i]:
            clip = batch["image"][i].to(model.device)
            outputs = preds
            clip_idx = i
            break
    if clip is not None:
        break
```

```python
os.makedirs("outputs", exist_ok=True)
video_path = "outputs/anomaly_clip.mp4"

anomaly_map = outputs.anomaly_map[clip_idx].cpu().numpy()  # (T, H, W)
pred_score = outputs.pred_score[clip_idx].item()
pred_label = outputs.pred_label[clip_idx].item()
```

```python
anomaly_map = (anomaly_map - anomaly_map.min()) / (anomaly_map.max() - anomaly_map.min() + 1e-8)
```

```python

frames = []
for t in range(clip.shape[0]):
    frame = clip[t].cpu().permute(1, 2, 0).numpy()
    amap = anomaly_map
    overlay = 0.6 * frame + 0.4 * plt.cm.jet(amap)[..., :3]
    overlay = np.clip(overlay, 0, 1)
    overlay = np.squeeze(overlay)

    # Convertir a uint8 [0–255]
    frame_uint8 = (overlay * 255).astype(np.uint8)
    frames.append(frame_uint8)

# Guardar video con códec explícito
with imageio.get_writer(video_path, fps=5, codec="libx264", format="FFMPEG") as writer:
    for frame in frames:
        writer.append_data(frame)

print(f" Video guardado correctamente en: {video_path}")
```

```python
best_clip = None
best_outputs = None
best_score = -1.0
best_idx = (None, None)

# === Buscar el clip con mayor pred_score ===
for batch_idx, batch in enumerate(datamodule.test_dataloader()):
    with torch.no_grad():
        preds = model(batch["image"].to(model.device))
    
    # Obtener los scores del batch actual
    scores = preds.pred_score.cpu().numpy()
    max_idx = np.argmax(scores)
    max_score = scores[max_idx]
    
    if max_score > best_score:
        best_score = max_score
        best_clip = batch["image"][max_idx].to(model.device)
        best_outputs = preds
        best_idx = (batch_idx, max_idx)

print(f"Clip más anómalo encontrado: batch={best_idx[0]}, idx={best_idx[1]}, score={best_score:.3f}")

# === Extraer mapa de anomalía (2D) ===
```

```python
# === Extraer mapa de anomalía (2D) ===
anomaly_map = best_outputs.anomaly_map[best_idx[1]].squeeze().cpu().numpy()  # (H, W)
pred_label = best_outputs.pred_label[best_idx[1]].item()

# Normalizar mapa
anomaly_map = (anomaly_map - anomaly_map.min()) / (anomaly_map.max() - anomaly_map.min() + 1e-8)
```

```python
# === Crear video ===
frames = []
for t in range(best_clip.shape[0]):
    frame = best_clip[t].cpu().permute(1, 2, 0).numpy()
    amap = anomaly_map
    overlay = 0.6 * frame + 0.4 * plt.cm.jet(amap)[..., :3]
    overlay = np.clip(overlay, 0, 1)
    overlay = np.squeeze(overlay)
    frame_uint8 = (overlay * 255).astype(np.uint8)
    frames.append(frame_uint8)

video_path = "outputs/anomaly_clip_topscore.mp4"
with imageio.get_writer(video_path, fps=5, codec="libx264", format="FFMPEG") as writer:
    for frame in frames:
        writer.append_data(frame)

print(f"Video del clip más anómalo guardado en: {video_path}")
```

```python
import torch, numpy as np, matplotlib.pyplot as plt

datamodule.setup(stage="test")
model = engine.model
model.eval()

N_SAMPLES = 20
imgs = []
titles = []

with torch.no_grad():
    for batch in datamodule.test_dataloader():
        preds = model(batch["image"].to(model.device))
        B = len(preds.pred_score)
        for i in range(B):
            clip = batch["image"][i].to(model.device)          # (T,C,H,W)
            frame = clip[clip.shape[0]//2].cpu().permute(1,2,0).numpy()
            amap = preds.anomaly_map[i].squeeze().cpu().numpy()
            # normalizar mapa
            amap = (amap - amap.min()) / (amap.max() - amap.min() + 1e-8)
            overlay = np.clip(0.6*frame + 0.4*plt.cm.jet(amap)[..., :3], 0, 1)

            imgs.append(overlay)
            titles.append(f"score={preds.pred_score[i].item():.3f}, label={int(preds.pred_label[i].item())}")
            if len(imgs) >= N_SAMPLES:
                break
        if len(imgs) >= N_SAMPLES:
            break

cols = 5
rows = int(np.ceil(len(imgs)/cols))
fig, axes = plt.subplots(rows, cols, figsize=(3*cols, 3*rows))
axes = axes.ravel()
for ax, img, t in zip(axes, imgs, titles):
    ax.imshow(img)
    ax.set_title(t, fontsize=8)
    ax.axis("off")
for ax in axes[len(imgs):]:
    ax.axis("off")
plt.tight_layout()
plt.show()
```
